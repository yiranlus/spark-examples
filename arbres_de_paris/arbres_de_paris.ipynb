```python
import pyspark.pandas as ps
import matplotlib.pyplot as plt
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("arbres_de_paris") \
    .config("spark.sql.ansi.enabled", "false") \
    .getOrCreate()
sc = spark.sparkContext
```

# Arbres de Paris Analysis with PySpark

```python
df_spark = ps.read_csv("arbresremarquablesparis.csv", sep=";")
df_spark.head()
```

```python
df_spark['Année de plantation'] = ps.to_numeric(df_spark['Année de plantation'], errors="coerce")
```

```python
print("Le plus vieil arbre a été planté en :", end=" ")
df_spark['Année de plantation'].min()
```

```python
print("Le plus grand arbre a une circonférence de :", end=" ")
df_spark['circonference en cm'].max()
```

```python
print("La hauteur moyenne des arbres est de :", end=" ")
df_spark['hauteur en m'].mean()
```

```python
print("L'arrondissement du plus grand arbre est le :", end=" ")
df_spark.loc[df_spark['circonference en cm'].argmax()]['arrondissement3']
```

```python
print("Nombre d'arbres situés au Cimetière du Père Lachaise :", end=" ")
len(df_spark.loc[df_spark['Site'] == "Cimetière du Père Lachaise"])
```
